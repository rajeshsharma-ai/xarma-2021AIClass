# Linear Algebra

Equation of a line: $y = intercept + slope*x$

Weighted sum: $y = (w_{11} * i_1) + (w_{12} * i_2)$, 
where, 
$w_{11}=slope, w_{12}=intercept, i_1=x, i_2=1$

Matrix: $y = \begin{pmatrix} i_1 & i_2 \end{pmatrix} . \begin{pmatrix} w_{11} \\ w_{12} \end{pmatrix}$

Do you understand the above linear algebra and the equivalence of the three equations?

No? You should take a quick refresher [here](https://www.khanacademy.org/math/linear-algebra/vectors-and-spaces#linear-combinations).
